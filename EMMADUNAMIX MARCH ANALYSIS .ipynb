```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Load data
file_path = 'C:/Users/hp/Downloads/EMMADUNAMIX MAINTENACE 2025 - FIRST QUARTER EMMA DUNAMIX MARCH 2026 - Copy.csv'
df = pd.read_csv(file_path)
df
```

```python
# Cleaning logic
# 1. Rename columns (stripping whitespace)
df.columns = [col.strip() if col.strip() != "" else "MONTH_LABEL" for col in df.columns]
df.columns
```

```python
#Clean numeric columns
def clean_numeric(val):
    if pd.isna(val) or str(val).strip() == '':
        return 0.0
    if isinstance(val, str):
        # Remove commas, any non-numeric except dot
        clean_val = "".join(c for c in val if c.isdigit() or c == '.')
        return float(clean_val) if clean_val else 0.0
    return float(val)

df['AMOUNT'] = df['AMOUNT'].apply(clean_numeric)
df['RATE'] = df['RATE'].apply(clean_numeric)
df['QTY'] = pd.to_numeric(df['QTY'], errors='coerce').fillna(0)
df
```

```python
df=df.drop('LIST OF SPARE PARTS', axis=1)
df
```

```python
# 3. Clean DATE
# Assuming format DD/MM/YYYY based on the snippet "02/03/2026"
# Note: "MARCH 02/03/2026" was shown in head, which means there's a prefix
def extract_date(val):
    if pd.isna(val): return None
    val = str(val).strip()
    # If the date starts with "MARCH", "FEBRUARY" etc followed by space
    parts = val.split()
    date_str = parts[-1] if parts else val
    return date_str

df['DATE_CLEAN'] = df['DATE'].apply(extract_date)
df['DATE_CLEAN'] = pd.to_datetime(df['DATE_CLEAN'], format='%d/%m/%Y', errors='coerce')
df

```

## Truck Usage & Cost Summary Analysis

## Objective

This analysis identifies the **top 10 most frequently used trucks** and evaluates their **total expenditure** within the dataset.


## Methodology

The analysis was performed using the following steps:

1. **Grouping Data**

   * The dataset was grouped by **Truck Number** to consolidate all records related to each truck.

2. **Aggregation**

   * Two key metrics were calculated:

     * **Total Cost**: Sum of all expenses associated with each truck
     * **Frequency**: Number of times each truck appears in the dataset (indicating usage level)

3. **Sorting**

   * Results were sorted by **frequency (descending)** to highlight the most actively used trucks.

4. **Top Selection**

   * The **top 10 trucks** were selected for focused analysis.

5. **Data Formatting**

   * Total cost values were rounded and converted to **whole numbers** for clarity and readability.

```python
#Group by item to get count and sum
top_10 = df.groupby('ITEM PURCHASED')['AMOUNT'].agg(['count', 'sum']).reset_index()

# Rename columns and sort
top_10.columns = ['Item Name', 'Times Purchased', 'Total Amount']
top_10 = top_10.sort_values(by='Total Amount', ascending=False).head(30)

# Format the amount with commas
top_10['Total Amount'] = top_10['Total Amount'].apply(lambda x: f"{x:,.0f}")

print(top_10.to_string(index=False))
```

```python
# Grouping
repeated_items = df.groupby(
    ['TRUCK DRIVER', 'TRUCK NUMBER', 'ITEM PURCHASED']
).agg(
    How_Many_Times=('ITEM PURCHASED', 'count'),
    Total_Amount=('AMOUNT', 'sum'),
    Dates=('DATE', lambda x: ', '.join(x.dropna().astype(str).unique()))
).reset_index()

# Filter + Sort (best practice)
multi_items = repeated_items[
    repeated_items['How_Many_Times'] > 1
].sort_values(
    by=['How_Many_Times', 'Total_Amount'],
    ascending=[False, False]
)

# Format
multi_items_formatted = multi_items.head(50).copy()

multi_items_formatted['Total_Amount'] = (
    multi_items_formatted['Total_Amount']
    .round()
    .astype(int)
    .apply(lambda x: f"₦{x:,}")
)

print("REPEATED PURCHASES (SAME ITEM, SAME DRIVER, SAME TRUCK):")
print("-" * 100)
print(multi_items_formatted.to_string(index=False))
```
